```python
df4=df2.groupby('customer_id')['order_date'].count()
```

```python
df4
```

```python
df4=df4.reset_index()
```

```python
df4
```

```python
recency=df4.groupby('order_date')['customer_id'].count().reset_index()
```

```python
recency
```

```python
recency.rename(columns={'order_date':'recency'},inplace=True)
plt.pie(recency['customer_id'],labels=recency['recency'])
plt.title("Recency by customer_id")
plt.show
```

```python
df5=df2.groupby('customer_id')['order_id'].count()
```

```python
df5
```

```python
df5=df5.reset_index()
```

```python
df5
```

```python
Frequency=df5.groupby('order_id')['customer_id'].count().reset_index()
```
